```python

bulanik_cent = cv2.GaussianBlur(rgb, (33,33), 0)
plt.imshow(bulanik_cent)
plt.axis("off")
```

```python
_, binary_cent = cv2.threshold(gri_cent, 200, 200, cv2.THRESH_BINARY)
plt.imshow(binary_cent, cmap="gray")
plt.axis("off")
```

```python
kenar_cent = cv2.Canny(binary_cent, 127, 255)
plt.imshow(kenar_cent, cmap="gray")
plt.axis("off")
```

```python
contours, hierarchy = cv2.findContours(kenar_cent,  cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
kontorlu_cent = cv2.drawContours(rgb, contours, -1, (255,0,0), 4)

for i, kontor in enumerate(contours,start=1):
  if cv2.contourArea(kontor) > 180:
    x, y, w, h = cv2.boundingRect(kontor)
    cv2.rectangle(kontorlu_cent, (x,y), (x+w, y+h), (0,255,0), 4)
    cv2.putText(kontorlu_cent, str(i), (x,y), cv2.FONT_HERSHEY_COMPLEX, 3, (0,0,255), 5)
    cv2.putText(kontorlu_cent, "M.A", (x+20,y+60), cv2.FONT_HERSHEY_COMPLEX, 1, (0,0,255), 5)
    print(cv2.contourArea(kontor))
plt.imshow(kontorlu_cent)
```
